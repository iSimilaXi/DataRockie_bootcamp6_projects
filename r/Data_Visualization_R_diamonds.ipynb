```r
set.seed(33)
sample_diamonds <- sample_n(diamonds, 1000)
```

Most popular color of diamonds

```r
diamonds %>%
  ggplot(mapping = aes(x = color, fill = color)) +
  geom_bar() +
  labs(
    title = "Most popular color, from best [D] to worst [J]",
    x = "Color",
    caption = "Source: ggplot package") +
  theme_minimal() +
  scale_fill_brewer(type = "qual", palette = "Accent")
```

proportion of cut by color

```r
sample_diamonds %>%
  ggplot(mapping = aes(cut, fill = color)) +
  geom_bar(position = "fill") +
  labs(
    title = "Proportion of cut by color",
    x = "cut",
    caption = "source: ggplot package"
  ) +
  theme_minimal() +
  scale_fill_brewer(type = "qual", palette = "Accent")
```

relation between carat and price

```r
sample_diamonds %>%
  ggplot(mapping = aes(carat, price, color = cut)) +
  geom_point(size = 3, alpha = .7) +
  geom_smooth(method = "lm", color = "salmon", se = F) +
  labs(
    title = "Relationship between Carat and Price (USD)",
    x = "Carat",
    y = "Price (USD)",
    caption = "Source: ggplot package"
  ) +
  theme_minimal() +
  scale_color_brewer(type = "qual", palette = "Dark2")
```

boxplot of price (USD) by cut

```r
sample_diamonds %>%
  ggplot(mapping = aes(cut, price, color = cut)) +
  geom_boxplot(alpha = .7) +
  labs(
    title = "boxplot of price (USD) by cut",
    x = "Cut",
    y = "Price (USD)",
    caption = "Source: ggplot package"
  )
```

violin plot of carat by cut

```r
sample_diamonds %>%
  ggplot(mapping = aes(cut, carat, color = cut)) +
  geom_violin() +
  geom_jitter(width = .3, alpha = .2) +
  labs(
    title = "Violin plot of Carat by Cut",
    x = "Cut",
    y = "Carat",
    caption = "Source:ggplot package"
  ) +
  theme_minimal() +
  scale_color_brewer(type = "qual", palette = "Dark2")
```
