Python ATM (OOP stuff)
-create ATM class with 5 methods/functions

```python
class ATM():
  def __init__(self, name, balance, PIN):
      self.name = name
      self.balance = balance
      self.PIN = PIN

  def deposit(self, money):
      self.balance += money
      print(f"You've just deposited {money} Baht to your account.")
      print(f"New balance: {self.balance} Baht")

  def withdraw(self, money):
      if money > self.balance:
          print(f"Your balance: {self.balance} Baht")
          print("Your account has insufficient funds for this transaction.")
      else:
          self.balance -= money
          print(f"You've successfully withdrawn {money} Baht")
          print(f"New balance: {self.balance} Baht")

  def fast_cash(self):
      print('\nPlease select the amount of cash you want to withdraw.')
      print("(1) 100 Baht\n(2) 300 Baht\n(3) 500 Baht\n (4) 1000 Baht")
      num = int(input("Enter a number between 1 and 4: "))
      fcash = 0
      if num == 1:
          fcash = 100
          if fcash > self.balance:
              print(f"Your balance: {self.balance} Baht")
              print("Your account has insufficient funds for this transaction.")
          else:
              self.balance -= fcash
              print(f"You've successfully withdrawn {fcash} Baht")
              print(f"New balance: {self.balance} Baht")
      elif num == 2:
            fcash = 300
            if fcash > self.balance:
              print(f"Your balance: {self.balance} Baht")
              print("Your account has insufficient funds for this transaction.")
            else:
              self.balance -= fcash
              print(f"You've successfully withdrawn {fcash} Baht")
              print(f"New balance: {self.balance} Baht")
      elif num == 3:
            fcash = 500
            if fcash > self.balance:
              print(f"Your balance: {self.balance} Baht")
              print("Your account has insufficient funds for this transaction.")
            else:
              self.balance -= fcash
              print(f"You've successfully withdrawn {fcash} Baht")
              print(f"New balance: {self.balance} Baht")
      elif num == 4:
            fcash = 1000
            if fcash > self.balance:
              print(f"Your balance: {self.balance} Baht")
              print("Your account has insufficient funds for this transaction.")
            else:
              self.balance -= fcash
              print(f"You've successfully withdrawn {fcash} Baht")
              print(f"New balance: {self.balance} Baht")
      else:
          print(f'\n{num} is an invalid input.')

  def transfer(self, receiver_acc_name, money):
      if money > self.balance:
          print(f"Your balance: {self.balance} Baht")
          print("Your account has insufficient funds for this transaction.")
      else:
          self.balance -= money
          print("Transfer successful.")
          print(f"{money} Baht has been transferred to  {receiver_acc_name}.")
          print(f"New balance: {self.balance} Baht.")

  def change_PIN(self):
      new_PIN = input("\nEnter New PIN: ")
      reenter_new_PIN = input("Reenter New PIN")
      if new_PIN == reenter_new_PIN:
          self.PIN == new_PIN
          print("\nYour PIN has been changed.")
      else:
          print("\nNew PIN does not match.")
          print("\nPIN change failed.")
```

```python
#OOP inheritance
class KTB(ATM):
    def __init__(self, name, balance, PIN):
        super().__init__(name, balance, PIN)

    def change_acc_name(self, new_acc_name):
        self.name = new_acc_name
        print(f"New account name: {self.name}")
        print("Your accoubt name has been changed.")

    def balance_inquiry(self):
        print(f"\nAccoubt: {self.name}, Balance: {self.balance} Baht")
```

```python
# open an account
print("KTB Bank ready to your services.")
print("Open a bank account")
name = input("Enter your name here: ")
balance = int(input("Opening balance: "))
temp_PIN = '1543535'
customer = KTB(name, balance, temp_PIN)
print("\nYou've just open a KTB bank account!")
print(f"Your temporary PIN will be {temp_PIN}")
```

```python
# at ATM
print("Insert your credit card into an ATM machine.\n")

#PIN validation
retries = 3
while retries > 0:
    entered_PIN = input("Enter PIN: ")
    if entered_PIN != customer.PIN:
      print("Incorrect PIN.")
      retries -= 1
      if retries > 0:
        print(f"You have {retries} retries left.\n")
      else:
        print("The card will be retained.")
        break
    else:
        retries = 0
        #select transaction
        print("Please select transaction.")
        print("(1) Deposit, (2) Withdrawal, (3) Fast Cash, (4) Transfer, (5) PIN change, (6) Account Settings.")
        tx = int(input("Enter number here: "))
        if tx == 1:
            deposit = int(input("\nEnter the amount of cash you want to deposit."))
            customer.deposit(deposit)
        elif tx == 2:
            withdrawal = int(input("\nEnter the amount of cash you want to withdraw."))
            customer.withdrawal(withdrawal)
        elif tx == 3:
            customer.fast_cash()
        elif tx == 4:
            dest_acc = input("\nEnter the destination bank account.")
            transfer = int(input("\nEnter the amount of cash you want to transfer."))
            customer.transfer(dest_acc, transfer)
        elif tx == 5:
            customer.change_PIN()
        elif tx == 6:
            print("\nAccount Settings")
            print("(1) Change account name (2) Balance inquiry")
            acc_set = int(input("\nEnter a number."))
            if acc_set == 1:
                new_name = input("\nPlease enter a new account name.")
                customer.change_acc_name(new_name)
            elif acc_set == 2:
                customer.balance_inquiry()
            else:
                print(f'\n{acc_set} is an invalid input.')
        else:
          print(f'\n{tx} is an invalid input.')
```
