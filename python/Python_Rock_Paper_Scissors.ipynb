Rock Paper Scissors (w/ Python)

```python
import random as rd
```

```python
def RPS_Game() -> None:
  #greetings paragraph
    print("Rock/Paper/Scissors!")
    name = input("Enter name here: ")
    print(f"Challenger {name} dares to battle.")
    print("Get ready!")

  #counter
    count_round: int = 0
    count_win: int = 0
    count_lose: int = 0
    count_draw: int = 0

  #empty variable (string) used to determine player's choice
    player = ""

  #loop (like in R)
    while player != 'Done.':

        #count the number of rounds
        count_round += 1

        #bot choice
        bot = rd.choice(['Rock', 'Paper', 'Scissors'])

        #player's choices
        try:
            number = int(input("\nChoose wisely! 1:Rock, 2:Paper, 3:Scissors, 4:Check the score, 5:Done."))
            if number == 1:
              player = 'Rock'
            elif number == 2:
              player = 'Paper'
            elif number == 3:
              player = 'Scissors'
            elif number == 4:
              player = 'Check the score'
            elif number == 5:
              player == 'Done.'
            else:
              player = 'Wrong number, try again.'
              print(f'{number} is invalid, try again.')
              count_round -= 1
        except ValueError:
              print('Invalid input, try again.')
              player = 'Continue'
              count_round -= 1


        # scoring
        if    ((player == 'Rock'    and bot == 'Scissors') or\
              (player == 'Paper'    and bot == 'Rock') or\
              (player == 'Scissors' and bot == 'Paper')):
              count_win += 1
        elif  player == bot:
              count_draw += 1
        elif  ((player == 'Rock'     and bot == 'Paper') or\
               (player == 'Paper'    and bot == 'Scissors') or\
               (player == 'Scissors' and bot == 'Rock')):
               count_lose += 1

        #summary
        if  player in ['Rock', 'Paper', 'Scissors']:
            print(f"Round: {count_round} -> {name}: {player} vs. Bot: {bot}")
        elif player == 'Check the score':
          count_round -= 1
          print('Score Summary')
          print(f"Total Rounds: {count_round}")
          print(f"Wins: {count_win}, Draws: {count_draw}, Loses: {count_lose}")
        elif player == 'Done.':
          count_round -= 1
          print("Game's closing.")
          print(f"Total Rounds: {count_round}")
          print(f"Wins: {count_win}, Draws: {count_draw}, Loses: {count_lose}")
```
